```python
import pandas as pd
import seaborn as sns
import matplotlib
import numpy as np
```

```python
# Configure plots
matplotlib.style.use('seaborn-poster')  # Poster = big, paper = small
matplotlib.style.use('ggplot')
matplotlib.rcParams['font.family'] = 'serif'
sns.set_context('poster')   # Poster = big, paper = small
```

```python
# Read metadata file
metadata = pd.read_json("train_sample_videos/metadata.json")
```

```python
# Plot count of real vs false
real = metadata.loc["label"] == "REAL"
x = [False, True]
values = real.value_counts().to_numpy()
sns.barplot(x=x, y=values).set(title="Count", ylabel="Count")
```

```python
# Plot as percentage
values = np.divide(values, values.sum()) * 100
x = [False, True]
sns.barplot(x=x, y=values).set(title="Percentage", ylabel="Percentage")
```
